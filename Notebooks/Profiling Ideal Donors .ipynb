### Clustering Frequent Donors

#### Prepare Data for Clustering

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt


# Select relevant features
features = ["Age", "Genre_", "Arrondissement_de_résidence_", "A-t-il_(elle)_déjà_donné_le_sang_"] 
clustering_data = successful_donors[features].copy()


le_location = LabelEncoder()
clustering_data["Arrondissement_de_résidence_"] = le_location.fit_transform(clustering_data["Arrondissement_de_résidence_"])


# Normalize numerical features
scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data)
```

#### Determine the Optimal Number of Clusters (Elbow Method)

```python
# Finding optimal K using Elbow Method
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(clustering_data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='teal')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal K")
plt.show()
```

using the elbow method we can identify 4 clusters

#### Apply K-Means Clustering

```python
# Choose optimal K (based on elbow method, e.g., 3 clusters)
optimal_k = 3  # Adjust this based on elbow graph result
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(clustering_data_scaled)

# Add cluster labels to the dataframe
successful_donors["Cluster"] = clusters
```

#### Analyze Cluster Characteristics

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=successful_donors, x="Age", y="A-t-il_(elle)_déjà_donné_le_sang_", hue="Cluster", palette="viridis")
plt.xlabel("Age")
plt.ylabel("Past Donation History (0 = No, 1 = Yes)")
plt.title("Clustering of Donors Based on Age & Past Donation History")
plt.show()
```

If you count well the group of points youll see that there are 4. which supports what was derived above from the elbow method
