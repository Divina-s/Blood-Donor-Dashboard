In the context of blood donation, the letters "F" and "B" typically refer to the following types:

- **F (Full Blood Donation)**: This type involves the collection of whole blood from a donor, which can later be separated into its components (red blood cells, plasma, platelets).

- **B (Blood Component Donation)**: This refers to the donation of specific blood components rather than whole blood. Common component donations include:
  - **Platelet Donation**
  - **Plasma Donation**


performing **Mode Imputation** since it is simple and effective for skewed distributions. Since "F" represents about 90% of the data, using mode imputation ensures that most missing values are filled with the most common value, which is logical in this context.

```python
# Impute missing values with the mode
mode_value = df2['Type de donation '].mode()[0]  # Get the mode
df2['Type de donation ']=  df2['Type de donation '].fillna(mode_value)

# Verify the changes
print(df2['Type de donation '].value_counts())
```
